```python
tempTargetFilename = ['./data/indo-song-emot.csv', './data/augmented.csv']
targetFilename = tempTargetFilename[0]
resultFilename = ['./data/train-data.csv', './data/test-data.csv'] # Index 0 untuk train, 1 untuk test

emotion_label = ['bahagia', 'sedih', 'marah', 'takut']

# translator = GoogleTranslator(source='id', target='en')

base_dataset = pd.read_csv(targetFilename, index_col=0) #mengambil csv menjadi pandas, dan mendefinisikan kolom pertama sebagai index
column_csvs = base_dataset.columns.tolist() # simpen nama kolom untuk training dan testing nanti
base_dataset.head()
```

### 3.2 Menghitung Distribusi dari Tiap Label Emosi

```python
labels_count = base_dataset['emotion'].value_counts()
print(labels_count)
```

```python
read_df = pd.read_csv(resultFilename[0], encoding='utf-8')
dict_term = defaultdict(set) # {emo: setof term}
all_word = set()

for idx, row in read_df.iterrows():
    emo = row['emotion']
    full = row['full']
    song = row['title']

    full = preProcess(full)
    full = set(full)
    all_word.update(full)

    dict_term[emo].update(full)

dict_wordCountInLabel = defaultdict(list)

for key in dict_term.keys():
    print(key, dict_term[key])

for word in all_word:
    for emo in emotion_label:
        if word in dict_term[emo]:
            dict_wordCountInLabel[word].append(emo)

for word in dict_wordCountInLabel:
    print(word, dict_wordCountInLabel[word])
```

### Augmenting

```python
import pandas as pd
import random as rd

def randMeUp(df_base, dc_t, dwcil, emot):
    global labels_count
    luckyNumba = rd.randint(0, labels_count[emot] - 1)
    randRow = df_base.iloc[luckyNumba]

    full_size = len(randRow['full']) 
    reff_size = len(randRow['reff']) 

    if full_size > 10 and reff_size > 4:
        full_size += rd.randint(-9, 2)
        reff_size += rd.randint(-3, 4)

    temp_full_ly = []
    i = 0
    while i < full_size:
        luckyTerm = rd.sample(list(dc_t[emot]), 1)
        luckyTerm = str(luckyTerm[0])
        if len(dwcil[luckyTerm]) < rd.randint(2,5):
            temp_full_ly.append(luckyTerm)
            i += 1
    
    temp_reff_ly = []
    i = 0
    while i < reff_size:
        luckyTerm = rd.sample(list(dc_t[emot]), 1)
        luckyTerm = str(luckyTerm[0])
        if len(dwcil[luckyTerm]) < rd.randint(2,5):
            temp_reff_ly.append(luckyTerm)
            i += 1
            # print(luckyTerm)
    
    temp_full_ly = ' '.join(temp_full_ly)
    temp_reff_ly = ' '.join(temp_reff_ly)

    return temp_full_ly, temp_reff_ly
    

max_count = labels_count.max()

# This will hold all the new augmented rows
list_of_new_augmented_rows = []
percentage = 0.05

print(f"Target count for each class: {max_count}")

for emo in emotion_label:
    current_count = labels_count.get(emo, 0) # Get current count, default to 0 if label not in counts
    rows_needed_to_generate = int((max_count - current_count) * percentage) 

    print(f"\nProcessing emotion: '{emo}'")
    print(f"  Current count: {current_count}, Rows needed to generate: {rows_needed_to_generate}")

    if rows_needed_to_generate > 0:
        original_samples_for_emo = base_dataset[base_dataset['emotion'] == emo]
        num_original_for_emo = len(original_samples_for_emo)

        if num_original_for_emo == 0 and rows_needed_to_generate > 0:
            print(f"  Warning: No original samples found for '{emo}' to base augmentation on. Skipping this emotion.")
            continue

        # print("dc_t: ",dict_term)
        # print("dwcil: ", dict_wordCountInLabel)

        for i in range(rows_needed_to_generate):
            title = f"Augmented Title - {emo} {i+1}"
            artist = "Augmented Artist"
            emotion_val = emo 

            augmented_full_text, augmented_reff_text = randMeUp(original_samples_for_emo, dict_term, dict_wordCountInLabel, emo)
            # print(augmented_full_text)
            # print(augmented_reff_text)
            
            new_row = {
                'title': title,
                'artist': artist,
                'full': augmented_full_text,
                'reff': augmented_reff_text,
                'emotion': emotion_val
            }
            list_of_new_augmented_rows.append(new_row)


        print(f"  Generated {rows_needed_to_generate} new rows for '{emo}'.")

augmented_df = pd.DataFrame(list_of_new_augmented_rows)
augmented_df = pd.concat([augmented_df, base_dataset], axis=0).reset_index(drop=True)

augmented_df.to_csv("./data/augmented.csv", index=True)
```
